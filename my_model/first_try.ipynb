```python
# jsonからdfを作成
def make_df(file_path, min):
    price = []
    file = open(file_path, 'r', encoding='utf-8')
    data = json.load(file)
    if data['result'][str(min)]:
        for i in data['result'][str(min)]:
            if i[1] and i[2] and i[3] and i[4]:
                price.append({
                    'close_time': i[0],
                    'close_time_dt': datetime.fromtimestamp(i[0]).strftime('%Y/%m/%d %H:%M'),
                    'open_price': i[1],
                    'high_price': i[2],
                    'low_price': i[3],
                    'close_price': i[4],
                })
        df = pd.json_normalize(price)
        return df
    else:
        print('データがありません')
        return
    
def 
```
